```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read the data from CSV
def process_and_visualize_data(file_path):
    df = pd.read_csv(file_path)
    
    # Create figures for each year
    years = df['Year'].unique()
    
    for year in years:
        year_data = df[df['Year'] == year].copy()
        
        # Set up the figure with subplots
        fig, axes = plt.subplots(2, 2, figsize=(20, 20))
        fig.suptitle(f'Company Metrics Analysis for Year {year}', fontsize=16, y=0.95)
        
        # 1. Top 5 Companies by Revenue
        top_revenue_data = year_data.nlargest(5, 'Revenue')
        sns.barplot(data=top_revenue_data, x='Company Name_x', y='Revenue', palette='viridis', ax=axes[0, 0])
        axes[0, 0].set_title(f'Top 5 Companies by Revenue - {year}')
        axes[0, 0].set_xticks(range(len(top_revenue_data)))
        axes[0, 0].set_xticklabels(top_revenue_data['Company Name_x'], rotation=45, ha='right')
        axes[0, 0].set_ylabel('Revenue (Billions USD)')
        
        # 2. Bottom 5 Companies by Revenue
        bottom_revenue_data = year_data.nsmallest(5, 'Revenue')
        sns.barplot(data=bottom_revenue_data, x='Company Name_x', y='Revenue', palette='viridis', ax=axes[0, 1])
        axes[0, 1].set_title(f'Bottom 5 Companies by Revenue - {year}')
        axes[0, 1].set_xticks(range(len(bottom_revenue_data)))
        axes[0, 1].set_xticklabels(bottom_revenue_data['Company Name_x'], rotation=45, ha='right')
        axes[0, 1].set_ylabel('Revenue (Billions USD)')
        
        # 3. Top 5 Companies by ESG Score
        top_esg_data = year_data.nlargest(5, 'esgScore')
        sns.barplot(data=top_esg_data, x='Company Name_x', y='esgScore', palette='magma', ax=axes[1, 0])
        axes[1, 0].set_title(f'Top 5 Companies by ESG Score - {year}')
        axes[1, 0].set_xticks(range(len(top_esg_data)))
        axes[1, 0].set_xticklabels(top_esg_data['Company Name_x'], rotation=45, ha='right')
        axes[1, 0].set_ylabel('ESG Score')
        
        # 4. Bottom 5 Companies by ESG Score
        bottom_esg_data = year_data.nsmallest(5, 'esgScore')
        sns.barplot(data=bottom_esg_data, x='Company Name_x', y='esgScore', palette='magma', ax=axes[1, 1])
        axes[1, 1].set_title(f'Bottom 5 Companies by ESG Score - {year}')
        axes[1, 1].set_xticks(range(len(bottom_esg_data)))
        axes[1, 1].set_xticklabels(bottom_esg_data['Company Name_x'], rotation=45, ha='right')
        axes[1, 1].set_ylabel('ESG Score')
        
        # Adjust layout
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()

# Example usage
file_path = '/Users/anjaliraj/Documents/Datathon/Merged.csv'  # Replace with your CSV file path
process_and_visualize_data(file_path)
```
